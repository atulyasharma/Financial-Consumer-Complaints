```python
import numpy as np
import pandas as pd
```

```python
df2 = pd.read_csv("Consumer_Complaints.csv")
```

```python
df2
```

```python
# Renaming Columns

df2.rename(columns = {'Date received' : 'Date_Received', 'Product' : 'Product', 'Sub-product' : 'Sub_Product' , 
                           'Issue' : 'Issue' , 'Sub-issue' : 'Sub_Issue' , 
                            'Consumer complaint narrative' : 'Consumer_Complaint_Narrative' , 
                           'Company public response' : 'Company_Public_Response' , 'Company' : 'Company' , 
                           'State' : 'State' , 'ZIP code' : 'Zip_Code' , 'Tags' : 'Tags' , 
                      'Consumer consent provided?' : 'Consumer_Consent_Provided' , 
                           'Submitted via' : 'Submitted_Via' , 'Date sent to company' : 'Data_Sent_To_Company' , 
                            'Company response to consumer' : 'Company_Response_To_Consumer' , 
                      'Timely response?' : 'Timely_Response' , 'Consumer disputed?' : 'Consumer_Disputed' , 
                           'Complaint ID' : 'Complaint_ID'}, inplace = True)
```

```python
df2
```

```python
df2.info()
```

```python
df2
```

```python
df3=df2
df3
```

```python
# Replacing Null Values with Other Values
df3.Sub_Product = df2.Sub_Product.fillna('Unknown')
df3.Sub_Issue = df2.Sub_Issue.fillna('Unknown')
df3.Consumer_Complaint_Narrative = df2.Consumer_Complaint_Narrative.fillna('Unknown')
df3.Company_Public_Response = df2.Company_Public_Response.fillna('Unknown')
df3.Tags = df2.Tags.fillna('No_Tags')
df3.Consumer_Consent_Provided = df2.Consumer_Consent_Provided.fillna('Unknown')
df3.Consumer_Disputed = df2.Consumer_Disputed.fillna('Unknown')
df3.State = df2.State.fillna('Unknown')
df3.Zip_Code = df2.Zip_Code.fillna('00000')
```

```python
df3['Date_Received'] = pd.to_datetime(df3['Date_Received'])
df3['Data_Sent_To_Company'] = pd.to_datetime(df3['Data_Sent_To_Company'])
df3['Zip_Code'] = df3['Zip_Code'].str.replace('X', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('-', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('`', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('[', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace(' ', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('+', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('.', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('/', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('$', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('!', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('*', '0')
df3['Zip_Code'] = df3['Zip_Code'].str.replace('(', '0')
df3['Product'] = df3['Product'].str.replace(' ', '_')
df3['Sub_Product'] = df3['Sub_Product'].str.replace(' ', '_')
df3['Issue'] = df3['Issue'].str.replace(' ', '_')
df3['Sub_Issue'] = df3['Sub_Issue'].str.replace(' ', '_')
df3['Consumer_Complaint_Narrative'] = df3['Consumer_Complaint_Narrative'].str.replace(' ', '_')
df3['Company_Public_Response'] = df3['Company_Public_Response'].str.replace(' ', '_')
df3['Company'] = df3['Company'].str.replace(' ', '_')
df3['Tags'] = df3['Tags'].str.replace(' ', '_')
df3['Consumer_Consent_Provided'] = df3['Consumer_Consent_Provided'].str.replace(' ', '_')
df3['Company_Response_To_Consumer'] = df3['Company_Response_To_Consumer'].str.replace(' ', '_')
df3
```

```python
df3.info()
```

```python
# Reordering of Columns
df4 = df3[['Complaint_ID', 'Company', 'Date_Received', 'Data_Sent_To_Company', 'Product', 'Sub_Product',
           'Issue', 'Sub_Issue', 'Submitted_Via', 'Consumer_Complaint_Narrative', 'State', 'Zip_Code', 'Tags',
           'Company_Public_Response', 'Company_Response_To_Consumer', 'Timely_Response', 'Consumer_Disputed',
           'Consumer_Consent_Provided']]
```

```python
df4
```

```python
df4['Zip_Code'] = df4['Zip_Code'].astype(int)
df4
```
